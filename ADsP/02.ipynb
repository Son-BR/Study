## 1. 데이터 분석 기획의 이해

========================================================================================================================

### 1.1 분석 기획 방향성 도출

#### 1, 분석 기획의 특징

 - 분석 기획 : 과제를 정의하고 결과를 도출 할 수 있도록 관리 방안을 사전에 계획하는 일련의 작업
 - 데이터 사이언티스트의 역량 : 수학/통계학적 지식, 정보기술, 비즈니스에 대한 이해와 전문성

#### 2. 분석 대상과 방법
 - 최적화(Optimization) : 분석 대상을 알고 있고, 분석 방법도 알고 있는 경우
 - 통찰(Insight) : 분석 방법과 분석 구조의 활용은 하고 있으나, 분석 대상이 무엇인지 인지하지 못하는 유형
 - 솔루션(Solution) : 분석 대상은 알고 있지만, 분석 방법을 모르는 경우
 - 발견(Discovery) : 분석의 대상과 분석방법을 모두 모르는 경우

 <img src="./Data/분석의대상과방법.png">
 
#### 3. 목표 시점별 분석기획 방안
 - 과제중심적 접근방식(당면한 분석주제의 해결)
   - 속도와 검사(speed & test)
   - 즉각적인 실행을 통한 성과 도출(Quick & Win)
   - 문제 해결(Problem Solving)
 - 마스터플랜 단위(지속적 분석 문화 내재화)
   - 정확도와 전개(Accuracy & Deploy)
   - 장기적인 관점(Long Term View)
   - 문제 정의(Problem Definition)

#### 4. 분석 기획 시 고려사항
 - 가용 데이터(Available Data)에 대한 고려
 - 적절한 유즈케이스(Proper Business Use Case) 탐색 필요
 - 사전계획 수립 필요(Low Barrier Of Execution)

### 1.2 분석 방법론

#### 1. 분석 방법론 개요
 - 기업의 합리적 의사결정을 가로막는 장애요소
   - 고정관념(Stereotype), 편향된 생각(Bias), 프레이밍 효과(Framing Effect)
 - 방법론의 적용 업무의 특성에 따른 모델
   - 폭포수 모델(Waterfall Model) : 단계를 순차적으로 진행, 문제발생시 이전단계로(피드백)
   - 애자일 모델(Agile Model) : 폭포수 모델 반대, 끊임없이 개발하고 수정(순환)
   - 프로토타입 모델(Prototype Model) : 폭포수 모델 개선, 미완성 모델을 고객에게 제공
   - 나선형 모델(Spiral Model) : 여러 개발과정을 점증적으로 진행, 처음시도하는 프로젝트에 용이, 복잡해질 가능성

#### 2. KDD 분석 방법론
 - 데이터셋 선택 -> 데이터 전처리 -> 데이터 변환 -> 데이터 마이닝 -> 결과 평가
 - 특징 : 5단계, 데이터 마이닝

#### 3. CRISP-DM 분석 방법론
 - 업무 이해 -> 데이터 이해 -> 데이터 준비 -> 모델링 -> 평가 -> 전개
 - 특징 : 6단계, 업무 이해 후 모델링

#### 4. 빅데이터 분석 방법론
 - 빅데이터 분석 방법론의 5단계
   - 분석기획 -> 데이터 준비 -> 데이터 분석 -> 시스템 구현 -> 평가 및 전개
   - 분석기획(Planning) : 문제점 인식, 수행계획 수립, 위험계획 수립, 프로젝트 범위 설정
   - 데이터 준비(Preparing) : 데이터 정의 및 준비, 데이터 수집
   - 데이터 분석(Analyzing) : 다양한 분석 기법과 알고리즘 이용
   - 시스템 구현(Developing) : 기획에 맞는 모델 도출, 운영 중인 시스템에 적용, 프로토타입 구현
   - 평가 및 전개(Deploying) : 모델 발전계획 수립, 프로젝트 평가 및 보고
   - 특징 : 5단계, 분석 기획 후 시스템 구현

- 빅데이터 분석의 계층적 프로세스
   - 단계(Phase) : 최상위 계층, 단계별 산출물 생성, 각 단계는 기준선으로 설정, 버전관리 등으로 통제
   - 테스크(Taks) : 단계를 구성하는 단위 활동, 물리적 또는 논리적 단위로 품질 검토
   - 스탭(Step) : 입력자료(Input), 처리 및 도구(Process&Tool), 출력자료(Output)로 구성된 단위 프로세스

### 1.3 분석 과제 발굴

#### 1. 분석과제 발굴 방법론
 - 하향식 접근 방식(Top Down Approach) : 분석과제가 주어지고 해법 탐색
 - 상향식 접근 방식(Bottom Up Approach) : 문제의 정의자체가 어려운 경우, 데이터 기반으로 문제점 개선

#### 2. 하향식 접근 방식(Top Down Approach)
 - 문제 탐색 -> 문제정의 -> 해결방안 탐색 -> 타당성 검토
 - 문제 탐색 : 업무, 제품, 고객, 규제와 감사, 지원 인프라의 영역으로 기업 비즈니스 분석
 - 문제 정의 : 비즈니스 문제를 데이터의 문제로 변환하여 정의
 - 해결방안 탐색 : 분석역량, 분석기법 및 시스템으로 해결방안 탐색
 - 타당성 검토 : 경제적 타당성, 데이터 및 기술적 타당성 검토

#### 3. 상향식 접근 방식(Bottom Up Approach)
 - 정의 : 데이터로 부터 분석을 통하여 통찰력과 지식을 얻는 접근 방법
 - 특징 : 다양한 환경에서 발생하는 문제들에 적용 가능, 데이터 그 자체를 객관적으로 관찰, 비지도 학습 방법, 시행착오를 통한 문제 해결(프로토 타이핑)

#### 4. 분석과제 정의
 - 분석 과제 정의서를 통해 분석별 필요 소스 데이터, 분석방법, 데이터 입수 및 분석의 난이도, 분석 수행주기, 검증 오너십, 상세 분석 과정 등을 정의

========================================================================================================================

### 1.4 분석 프로젝트 관리 방안

#### 1. 분석과제 관리를 위한 5가지 주요 영역
 - 분석프로젝트는 범위, 일정, 품질, 리스크, 의사소통 등 영역별 관리가 수행되어야 함
 - 5가지 주요 속성을 고려하여 추가적인 관리 필요
 - 데이터 크기, 데이터 복잡성, 분석 복잡성, 속도, 정확도&정밀도

#### 2. 분석 프로젝트의 특성
 - 분석가의 목표 : 개별적인 분석업무 수행 뿐만 아니라 전반적인 프로젝트 관리 또한 중요
 - 분석가의 입장 : 데이터 영역과 비즈니스 영역의 현황을 이해하고, 프로젝트의 목표인 분석의 정확도 달성과 결과에 대한 가치 이해를 전달하는 조정자로서의 분석가 역할이 중요
 - 분석 프로젝트는 도출된 결과의 재해성을 통한 지속적인 반복 및 정교화가 수행되는 경우가 대부분이므로 프로토 타이핑 방식의 애자일 프로젝트 관리방식에 대한 고려 필요

### 2.1 마스터 플랜 수립 프레임 워크

#### 1. 마스터 플랜 수립 프레임 워크
 - 분석 과제를 대상으로 다양한 기준을 고려해 우선순위를 설정하고, 데이터 분석 구현을 위한 로드맵을 수립
 - 우선순위 고려 요소 : 전략적 중요도, 비즈니스성과/투자회수율(ROI), 실행 용이성
 - 적용범위/방식 고려 요소 : 업무 내재화 적용 수준, 분석 데이터 적용 수준, 기술 적용 수준

#### 2. 투자회수율(ROI) 관점에서 빅데이터의 핵심 특징
 - 투자비용 요소
   - 3V(Volume, Variety, Velocity)
 - 비즈니스 효과 요소
   - Value(가치)

### 2.2 분석 거버넌스 체계 수립

#### 1. 분석 거버넌스 체계 구성요소
 - 지속적인 데이터 분석의 적용과 확산을 위한 거버넌스 체계
 - 데이터 -> 분석교육/마인드 육성체계 -> 분석기획 및 관리수행 조직 -> 과제 기획 및 운영 프로세스 -> 분석관련시스템 -> 데이터

#### 2. 데이터 분석 수준진단
 - 분석 준비도 : 분석업무, 인력/조직, 기법, 데이터, 분석문화, 분석인프라
 - 분석 성숙도 : 도입 -> 활용 -> 확산 -> 최적화 // 비즈니스, 조직 및 역량, IT

 <img src="./Data/준비도성숙도.png">

#### 3. 데이터 거버넌스 체계 수립
 - 데이터 거버넌스 개요
   - 전사 차원의 모든 데이터에 대하여 정책 및 지침, 표준화, 운영조직 및 책임 등의 표준화된 관리 체계를 수립하고 운영을 위한 프레임워크 및 저장소를 구축하는 것
   - 마스터 데이터, 메타 데이터, 데이터 사적은 데이터 거버넌스의 중요한 관리 대상
 - 데이터 거버넌스 구성요소
   - 원칙(Principle), 조직(Organization), 프로세스(Process)
 - 데이터 거버넌스 체계
   - 데이터 표준화, 데이터 관리 체계, 데이터 저장소 관리(Repository), 표준화 활동

#### 4. 데이터 분석을 위한 3가지 조직 구조 : 집중구조, 기능구조, 분산구조
 - 집중구조 : 별도의 분석전담 조직
 - 기능구조 : 일반적인 분석 수행 구조, 업무부서에서 각자 기능 수행
 - 분산구조 : 현업부서로 직접 배치, 신속한 액션 가능

#### 5. 분석과제 관리 프로세스
 - 과제 발굴 : 분석 아이디어 발굴 -> 분석과제 후보제안 -> 분석과제 확정
 - 과제 수행 : 팀구성 -> 분석과제 실행 -> 분석과제 진행관리 -> 결과공유/개선
